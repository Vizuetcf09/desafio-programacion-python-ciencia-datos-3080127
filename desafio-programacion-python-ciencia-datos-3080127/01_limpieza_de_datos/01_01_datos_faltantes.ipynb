# Limpieza de datos

La limpieza de datos es el proceso de encontrar y corregir los datos  inexistentes o incorrectos presentes en el conjunto d edatos. Para la limpieza de datos se puede usar la libreria de código abierto Pandas.

La limpieza de datos es importante ya que
los conjuntos de datos tendrán valores faltantes por distintas razones, por lo que es necesario saber tratar con ellos.

Es la practica de corregir o eliminar datos inexactos, dañados, mal formateados, duplicados o incompletos de un conjunto de datos.

## Datos faltantes


Ocurre cuando falata información disponible para una o más columnas o para caracteristicas de las filas.

Los datos faltantes son datos que:

    - No se recoliparon, aunque existan
    - Datos que nunca existieron

Ej. Encuesta sobre los intereses de usuarios

    -  Información faltante sobre los ingresos mensuales


### Qué son sol datos falyantes

Los datos faltantes son: 

    - Observaciones que no se registraron
    - Datos corruptos (Corrupción de datos)


En pandas los datos faltantes se representan de dos formas:

    - None: objeto único de Python que se usa para datos faltantes
    - NaN: (Not a Number) Valor especial de punto flotante reconocido por todos los sistemas utilizan la representación estándar de IEEE.

####  Importancia del manejo de datos faltanes

Sumanejo es impoetnate, ya que muchos algoritmos de aprendizaje automático no los admite, por lo que es necesario hacer una limpeza de los datos, para lograrlo, debemos:

    - Revisar los valores nulos dentro de las columnas
    - Revisar las filas que puedan tener muchos valores faltantes.

#### Rellenar los datos faltantes (imputación)

A veces los valores numéricos vienen incompletos, ya sea como NaN o None, cuando es el caso, podemos relizar la técnica llamada **imputación**, que no spermite cambiar lo svalores faltantes con valores sencibles.


#### Código | Desafío

Rellevar los datos faltantes de la columna edad con la media de la edad.

##### Desafío | Imputación de datos

```python
### Rellenar valores nulos con un valor específico
#### Detectar los datos nulos
df_drop_columns.isnull().sum()
```

```python
##### Eliminamos datos nulos de columnas con datos te tipo string
df_dropna_str_columns = df_drop_columns.dropna(subset=['Nombre', 'Sexo',  'Estado_Civil', 'Gerencia', 'Departamento'])
df_dropna_str_columns.isnull().sum()
```

```python
##### Imputación de datos nulos en columnas con datos de tipo numericos (int, float)
# 1. Definimos las columnas numéricas que queremos cambiar
columnas_num = ['Edad', 'Experiencia', 'Salario']

df_dropna_num_columns = df_dropna_str_columns.dropna(subset=columnas_num)
df_dropna_num_columns.head()
```

```python
# 3. Calculamos la media y rellenamos solo en esas columnas
dropna_num_columns = df_drop_columns[columnas_num]
df_dropna_num_columns[columnas_num] = df_dropna_num_columns[columnas_num].fillna(df_dropna_num_columns[columnas_num].mean())

###### Comprobación de la imputación de datos nulos en columnas con datos de tipo numericos (int, float)
df_dropna_num_columns.isnull().sum()
```
