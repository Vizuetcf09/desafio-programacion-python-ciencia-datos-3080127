## Correlaciṕon entre dos variables

### Apuntes

#### Correlación

Denota una asociación entre dos variables cuantitativas.

#### Correlación positiva

En la correlación positova, las variables se mueven en la misma dirección

#### Correlación negativa

Se lellama correlacióin positiva, cuando el valor de una variable aumenta, y el valor de la otra disminuye

#### Correlación neutral o de valor cero

En este tipo de variables, no existe una correlación, por lo tanto, el valor de una no afecta el valor de la otra, son independientes.

#### Coeficiente de correlación de Pearson

El coeficiente de correlación puede tomar cualquier valor entre el rango de `-1` y `1`.

  - Cuando es **igual** a `1`, es una **correlación lineal positiva perfecta**
  - Cuando es **mayor** a `0`, es una **correlación positiva**
  - Cuando es **igual** a `0`, es una **correlación neutra o sin relación (correlación)**
  - Si es **menor** que `0`, es una **correlación negativa**
  - Cundo es **igual** a `-1`, es una **correlacion lineal negativa perfecta**.

## Normalizar las columnas de un dataframe

### Normalización | Estandarización de datos

La normalización es una habilidad muy importante para cualquier analista o científico de datos, implica ajustar los datos, enteros que existen en diferentes escalas, a una escala común, que permita compararlos más facilmente.

#### Estandarización de datos

La estandarización de los datos facilita que un modelo de aprendizaje automático aprenda la relación entre las entradas sin que un valor domine solo por el tamaño del número.

Por ejemplo:

La altura y el peso tienen valores diferentes y escalas distintas, por lo que es necesario normalizarlas, para que tengan la misma escala, que nos permita compararlos más facilmente.

### Desafío
